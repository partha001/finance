```python
dftemp = pivotedAssets
dftemp["total"] = dftemp.groupby(level=0, axis=1).sum()
#dftemp["total"] = dftemp.T.groupby(level=0).sum()
dftemp
```

```python

fig = plt.subplots(figsize=(15, 7))
plt.xlabel('Date', fontsize=15)
plt.ylabel('INR', fontsize=15)
plt.title('assets-vs-time', fontsize=20)
plt.plot(dftemp.index, dftemp[('amount', 'zerodha_equity')], label= 'zerodha_equity')
plt.plot(dftemp.index, dftemp[('amount', 'angelone_equity')], label='angelone_equity')
plt.plot(dftemp.index, dftemp[('amount', 'zerodha_mf')], label='zerodha_mf')
plt.plot(dftemp.index, dftemp[('amount', 'groww_mf')], label='groww_mf')
plt.plot(dftemp.index, dftemp[('total')], label='total')
plt.legend()
plt.savefig('test.png')
plt.show()
plt.close()
```
